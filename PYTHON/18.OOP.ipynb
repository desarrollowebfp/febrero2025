## Programación Orientada a Objetos


```python
# Clase -> Class -> Modelo o plantilla
# Objeto -> Object -> La instancia o versión concreta generada mediante una clase

class Pelicula:
    """
    Clase que representa la 'plantilla' de una película
    """
    def __init__(self, titulo, director, anio, puntuacion):
        """
        Método constructor
        """
        self.titulo = titulo
        self.director = director
        self.anio = anio
        self.puntuacion = puntuacion
    """
    Aparte del método constructor (que es obligatorio por buenas prácticas), podemos crear nuestros propios métodos
    """
    def mostrar_info(self):
        print(f"{self.titulo} se lanzó en {self.anio}, y fue dirigida por {self.director}, con una puntuación en crítica de {self.puntuacion}")
        
    def inflar_puntuacion(self, puntos):
        self.puntuacion += puntos
        print("¡La puntuación ha sido modificada!")
        
    # Clase hija que va a heredar propiedades de la clase madre "Pelicula"
class Pelicula3D(Pelicula):
    def __init__(self, titulo, director, anio, puntuacion):
        # Llamamos al constructor de la clase madre -> Pelicula
        super().__init__(titulo, director, anio, puntuacion)
        
    def avisar_gafas(self):
        print(f"La película {self.titulo} necesita gafas 3D")
```

```python
# Creamos un nuevo objeto a partir de la clase Pelicula
pelicula_origen = Pelicula("Origen", "Christopher Nolan", 2012, 8.5)
pelicula_toy_story = Pelicula("Toy Story", "John Lasseter", 1995, 8.8)
# Si sacamos la película creada mediante la propiedad integrada __dict__ la podremos consultar como un diccionario
print(pelicula_origen.__dict__)
print(pelicula_origen.director)
pelicula_origen.mostrar_info()
pelicula_toy_story.mostrar_info()

pelicula_toy_story.inflar_puntuacion(0.5)
pelicula_toy_story.mostrar_info()
```

```python
pelicula_avatar = Pelicula3D("Avatar", "James Cameron", 2009, 7.8)
pelicula_avatar.mostrar_info()
pelicula_avatar.avisar_gafas()

pelicula_origen.mostrar_info()
```
