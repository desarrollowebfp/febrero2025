```python
                #parametro nombre que recibe por argumento -> "Ander" - "Victor" - "Daniel"
def saludar_nombre (nombre):
    print(f"Hola {nombre}!")
    
saludar_nombre("Ander")
saludar_nombre("Victor")
saludar_nombre("Daniel")
```

```python
def sumar_dos_numeros(numA, numB):
    total = numA + numB
    print(total)
    
sumar_dos_numeros(5,8)
```

```python

def multiplicar_por_dos(numero):
    total = numero * 2
    return total
    
veinticinco_por_dos = multiplicar_por_dos(25)
print(veinticinco_por_dos)
```

```python
def saludar_usuario(nombre_usuario="User"):
    print(f"Hola {nombre_usuario}")
    
saludar_usuario()
saludar_usuario("Paco")
```

```python
x = 10 # variable global

def mostrar_valor():
    x = 5 # variable local
    print("Valor dentro de la función", x)
    
mostrar_valor()

print("Valor fuera de la función", x)
```

```python
# Documentación de funciones
def sumar_numeros(numA, numB):
    """
    Suma dos números y devuelve el resultado

    Args:
        numA (int): Primer número de la suma
        numB (int): Segundo número de la suma

    Returns:
        int: El resultado de la suma de ambos números
    """
    return numA + numB

sumar_numeros()
```
