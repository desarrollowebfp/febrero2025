## Reduce


```python
from functools import reduce

# reduce(funcion, iterable, valor inicial)

numeros = [1,2,3,4]

suma = reduce(lambda acc, x: acc + x, numeros, 0)
print(suma)

# Inicio acc = 0
# Toma x = 1 -> acc = 0 + 1 = 1
# Toma x = 2 -> acc = 1 + 2 = 3
# Toma x = 3 -> acc = 3 + 3 = 6
# Toma x = 4 -> acc = 6 + 4 = 10
## Final acc = 10 -> lo retorna
```

```python
numeros = [1,2,3,4]

numerosAlCuadrado = reduce(lambda acc, x: acc + [x**2], numeros, [])
print(numerosAlCuadrado)
```
